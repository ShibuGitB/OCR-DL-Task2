Email detection from Image

```python
import cv2 

img=cv2.imread(r"C:\Users\shibu\OneDrive\Documents\Deep learning projects\ocr\Screenshot_20231213-094718_OneDrive.jpg") 
```

```python
import pytesseract 
pytesseract.pytesseract.tesseract_cmd=r"c:\Program Files\Tesseract-OCR\tesseract.exe" 
```

```python
data=pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT) 
data
```

```python
n=len(data["text"]) 
n
```

```python
import re 

reg="^[a-z0-9!#$%^&*]+[@][g][m][a][i][l][.][c][o][m]"
```

```python
for i in range (n) : 
    
    if re.fullmatch(reg,data["text"][i]) : 
        
        print (data["text"][i]) 
        
        x=data["left"][i] 
        y=data["top"][i] 
        w=data["width"][i] 
        h=data["height"][i] 
        
        cv2.rectangle(img,(x,y),(x+w,y+h),color=(0,0,0),thickness=3) 
        
cv2.imshow("email",img) 
cv2.waitKey() 
cv2.destroyAllWindows()
```

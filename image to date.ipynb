Date Detection from Image

```python
import cv2 

img=cv2.imread(r"C:\Users\shibu\OneDrive\Documents\Deep learning projects\ocr\WhatsApp Image 2024-03-14 at 11.52.14_feda0ba2.jpg") 
```

```python
import pytesseract 
pytesseract.pytesseract.tesseract_cmd=r"c:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
data=pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT) 
data
```

```python
n=len(data["text"]) 
n
```

```python
import re 

for i in range (n) : 
    
    reg="[0-9]{1,2}[/][0-9]{1,2}[/][0-9]{4}"
    # if re.search(r"([0-9]{1,2}[/][0-9]{1,2}[/][0-9]{4})",data["text"][i]) : 
    if re.fullmatch(reg,data["text"][i]) :
        
        print (data["text"][i]) 
        
        x=data["left"][i] 
        y=data["top"][i] 
        w=data["width"][i] 
        h=data["height"][i] 
        
        cv2.rectangle(img,(x,y),(x+w,y+h),color=(0,0,0),thickness=4) 
        
cv2.imshow("date",img) 
cv2.waitKey() 
cv2.destroyAllWindows() 
```
